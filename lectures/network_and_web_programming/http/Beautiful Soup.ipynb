```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://wit.pwr.edu.pl/wydzial/struktura-organizacyjna/katedry")
soup = BeautifulSoup(page.content)
```

```python
soup.prettify()
```

```python
soup.find('div', attrs={'class': 'credits'})
```

```python
for li in soup.find('ul').children:
    print(li.text)
```

```python
# .class - selects elements with a given class
# parent > child - selects all children within parent tag
# [attribute=value] - selects with given attribute value
soup.select('.text > ul > li > strong > a[target=_blank]')
```
